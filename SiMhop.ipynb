# Change single-hop file 

```
## q_id

## question (is possible = true)

## [answer] --> [answer] or answer

# single ---> type 

# [[answer ,0], [answer, 0] ---> supporting fact 

## title ---> top titles --> [top title]

## context ===> [['title , ['context']]

squad_list = []


typee = 'single'
level = 'easy'


for i in range(len(squad_df)):
  title = squad_df.iloc[i]['data']['title']

  for j in range(len(squad_df.iloc[i]['data']['paragraphs'])):
    # for eval set
    #context = squad_df.iloc[i]['data']['paragraphs'][j]['context']

    #for train set
    context_alone = squad_df.iloc[i]['data']['paragraphs'][j]['context']
    context = [ [title, [context_alone] ] ] 


    for p in range(len(squad_df.iloc[i]['data']['paragraphs'][j]['qas'])):
      if squad_df.iloc[i]['data']['paragraphs'][j]['qas'][p]['is_impossible']==False:
        question= squad_df.iloc[i]['data']['paragraphs'][j]['qas'][p]['question']
        id = squad_df.iloc[i]['data']['paragraphs'][j]['qas'][p]['id']
        answer = squad_df.iloc[i]['data']['paragraphs'][j]['qas'][p]['answers'][0]['text']
        
        #for eval set
        #squad_list.append([id, question, [answer], 'single', [[answer, 0], [answer, 0]] , [title] ])
        
        #for train set
        squad_list.append([id, answer, context, level, question, [[answer, 0], [answer, 0]] , typee  ] )
 
```
